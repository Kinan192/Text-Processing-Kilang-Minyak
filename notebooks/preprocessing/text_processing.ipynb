Cleaning Data

```python
import pandas as pd
```

```python
df = pd.read_csv('../../data/raw/kilang_minyak_data.csv')
df.head(10)
```

Check Missing Value

```python
df.isnull().sum()
```

```python

df = df.drop(['username', 'location', 'image_url', 'in_reply_to_screen_name'], axis=1)
df.isnull().sum()
```

Casefolding dataset

```python
df['full_text'] = df['full_text'].str.lower()
df['tweet_url'] = df['tweet_url'].str.lower()
df.head(10)
```

Hapus tag username untuk full text komentar

```python
import re

df['full_text'] = df['full_text'].apply(lambda x: re.sub(r'@\w+', '', str(x)))

df['full_text'] = df['full_text'].str.replace('\s', '', regex=True).str.strip()

df.head(10)
```
